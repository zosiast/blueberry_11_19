```python
import netCDF4 as nc
import pylab as plt
import numpy as np
import pandas as pd
from mpl_toolkits.basemap import Basemap, shiftgrid
import numpy.ma as MA
```

Get data and do conversions

```python
#extract data and separate variables
temp_data = nc.Dataset('/shared/netscratch/ptg21/UM_data/xltln/xltln_temp.nc')
temp = temp_data.variables['temp'][:]

cl_data = nc.Dataset('/shared/netscratch/ptg21/UM_data/xltln/xltln_cl.nc')
cl = cl_data.variables['cl']
print(cl)
#units kg
mass_data = nc.Dataset('/shared/netscratch/ptg21/UM_data/xltln/xltln_airmass.nc')
mass = mass_data.variables['airmass_atm']
#print(mass)

press_data = nc.Dataset('/shared/netscratch/ptg21/UM_data/xltln/xltln_p.nc')
press = press_data.variables['p']
alt = press_data.variables['model_level_number'][:]

lat = cl_data.variables['latitude'][:]
lon = cl_data.variables['longitude'][:]
#alt_cl = cl_data.variables['level_height'][:]
```

```python
#Convert Cl kg kg-1 into no. molecules per cm3
r_gas = 8.3145
n_a = 6.02e23
conv_1 = (28.97*n_a)/(35.5*r_gas*1e6)*(np.divide(press,temp))
cl_conc = np.multiply(cl,conv_1[:120,:,:,:])
```

```python
#calculate rate(constant)
kcl = 6.6e-12*np.exp(-1240/temp)
kcl_cl = np.multiply(kcl[:120,:,:,:],cl_conc)
```

Start of Lawrence plot

```python
zon_press = np.mean(press, axis=(0,3))[:25,:]
zon_cl = np.mean(cl_conc, axis=(0,3))[:25,:]
```

```python
#get zones by pressure
height_1 = MA.masked_less(zon_press, 75000)
height_2 = MA.masked_outside(zon_press, 75000, 50000)
height_3 = MA.masked_outside(zon_press, 50000,25000)
```

```python
lat_array = np.array([lat] * 25)

sec_1 = MA.masked_greater(lat_array,-30)
sec_2 = MA.masked_outside(lat_array, -30, 0)
sec_3 = MA.masked_outside(lat_array, 0, 30)
sec_4 = MA.masked_less(lat_array, 30)
```

```python
heights = [height_1.mask, height_2.mask, height_3.mask]
sections = [sec_1.mask, sec_2.mask, sec_3.mask, sec_4.mask]
conditions = []
for h in heights:
    for s in sections:
        conditions.append([h,s])
cond = np.asarray(conditions)
#print(cond.shape)
#print(cond[0,:,:,:])
```

```python
#loop over conditions to get mean for each area
law_cl = np.empty((25,73))
mean_vals = []
for i in range(12):
    values = MA.masked_where(np.logical_or(*cond[i,:,:,:]),zon_cl) #extract temp values in section
    mean = np.mean(values) #get mean over this section
    mean_vals.append(mean)
    law_cl = np.where(values.mask, law_cl, mean)

law_cl2 = MA.masked_where(zon_press < 25000, law_cl)
law_cl3 = MA.masked_where(zon_press < 50000, law_cl)
```

```python
#Make boundary for Lawrence plot
bound_750 = []
for i in range(73):
    for j in range(25):
        if zon_press[j,i] < 75000:
            bound_750.append(j+1)
            break
            
bound_500 = []          
for i in range(73):
    for j in range(25):
        if zon_press[j,i] < 50000:
            bound_500.append(j+1)
            break
```

```python
plt.figure(dpi=120)
plt.pcolormesh(lat,alt[:25],law_cl2, cmap='summer', vmin=0, vmax=3400)
plt.colorbar()

plt.yticks([])

plt.plot(lat,bound_750, 'k-', lw=1, ls='dotted')
plt.plot(lat,bound_500, 'k-', lw=1, ls='dotted')

plt.text(-60,5,f'{mean_vals[0]:.0f}')
plt.text(-24,6,f'{mean_vals[1]:.0f}')
plt.text(5,6,f'{mean_vals[2]:.0f}')
plt.text(52,5,f'{mean_vals[3]:.0f}')
plt.text(-65,13,f'{mean_vals[4]:.0f}')
plt.text(-24,14,f'{mean_vals[5]:.0f}')
plt.text(5,14,f'{mean_vals[6]:.0f}')
plt.text(52,13,f'{mean_vals[7]:.0f}')
plt.text(-70,19,f'{mean_vals[8]:.0f}')
plt.text(-24,20,f'{mean_vals[9]:.0f}')
plt.text(5,20,f'{mean_vals[10]:.0f}')
plt.text(52,19,f'{mean_vals[11]:.0f}')

#label pressures
plt.text(-117,2,'750 hPa',fontsize=8.5, weight='bold')
plt.text(-117,13,'500 hPa',fontsize=8.5, weight='bold')
plt.text(-117,20,'250 hPa',fontsize=8.5, weight='bold')

plt.axvline(x=0, linewidth=1, linestyle=(0,(1,2)), color='k')
plt.axvline(x=30, linewidth=1, linestyle=(0,(1,2)), color='k')
plt.axvline(x=-30, linewidth=1, linestyle=(0,(1,2)), color='k')

plt.xlabel('Latitude')
#plt.ylabel('Model height')
plt.title('Mean [Cl] by region (xltln) / molec cm-3')
```

```python
plt.figure(dpi=120)
plt.pcolormesh(lat,alt[:25],law_cl3, cmap='summer')
plt.colorbar()

plt.plot(lat,bound_750, 'k-', lw=1, ls='dotted')
plt.plot(lat,bound_500, 'k-', lw=1, ls='dotted')

plt.text(-60,5,f'{mean_vals[0]:.2f}')
plt.text(-24,6,f'{mean_vals[1]:.2f}')
plt.text(5,6,f'{mean_vals[2]:.2f}')
plt.text(52,5,f'{mean_vals[3]:.2f}')
plt.text(-65,13,f'{mean_vals[4]:.2f}')
plt.text(-24,14,f'{mean_vals[5]:.2f}')
plt.text(5,14,f'{mean_vals[6]:.2f}')
plt.text(52,13,f'{mean_vals[7]:.2f}')
plt.text(-70,19,f'{mean_vals[8]:.2f}')
plt.text(-24,20,f'{mean_vals[9]:.2f}')
plt.text(5,20,f'{mean_vals[10]:.2f}')
plt.text(52,19,f'{mean_vals[11]:.2f}')

plt.axvline(x=0, linewidth=1, linestyle='dotted', color='k')
plt.axvline(x=30, linewidth=1, linestyle='dotted', color='k')
plt.axvline(x=-30, linewidth=1, linestyle='dotted', color='k')


plt.xlabel('Latitude')
plt.ylabel('Model height')
plt.title('Mean [Cl] by region')
```
