```python
import numpy as np
import netCDF4 as nc
import numpy.ma as MA
import cf
import cfplot as cfp
import pylab as plt
```

```python
press_data = nc.Dataset('/shared/netscratch/ptg21/UM_data/xltln/xltln_p.nc')
press = press_data.variables['p'][:]

temp_data = nc.Dataset('/shared/netscratch/ptg21/UM_data/xltln/xltln_temp.nc')
temp = temp_data.variables['temp']
zon_temp = np.mean(temp, axis=(0,3))[:25,:]

lat = press_data.variables['latitude'][:]
lon = press_data.variables['longitude'][:]
alt = press_data.variables['model_level_number'][:]

zon_press = np.mean(press, axis=(0,3))[:25,:]

#get first time value
times = press_data.variables['time']
t1 = int(times[0])
```

Separate by pressure to get three different height sections

```python
#get zones by pressure
height_1 = MA.masked_less(zon_press, 75000)
height_2 = MA.masked_outside(zon_press, 75000, 50000)
height_3 = MA.masked_outside(zon_press, 50000,25000)

cfp.gopen(rows=1, columns=3, bottom=0.2)
cfp.gpos(1)
cfp.con(f=height_1, x=lat, y=alt[:25], ptype=2,lines=False)
cfp.gpos(2)
cfp.con(f=height_2, x=lat, y=alt[:25], ptype=2,lines=False)
cfp.gpos(3)
cfp.con(f=height_3, x=lat, y=alt[:25], ptype=2,lines=False)
cfp.gclose()
```

```python
#make temp masked array for each level
h1_temp = MA.masked_where(height_1.mask, zon_temp)
h2_temp = MA.masked_where(height_2.mask, zon_temp)
h3_temp = MA.masked_where(height_3.mask, zon_temp)

#average over level
l1_t_mean = np.mean(h1_temp)
l2_t_mean = np.mean(h2_temp)
l3_t_mean = np.mean(h3_temp)

#add to array
zon_mean_temps = np.zeros((60, 73))
zon_mean_temps = np.where(level_1.mask, zon_mean_temps, l1_t_mean)
zon_mean_temps = np.where(level_2.mask, zon_mean_temps, l2_t_mean)
zon_mean_temps = np.where(level_3.mask, zon_mean_temps, l3_t_mean)

zon_mean_temps2 = MA.masked_where(zon_mean_temps[:25,:] == 0, zon_mean_temps[:25,:].astype(int))

plt.pcolormesh(lat,alt[:25],zon_mean_temps2)
#average over axes 0:time and 3:longitude to get zonal mean
plt.colorbar()

plt.xlabel('Latitude')
plt.ylabel('Model height')
plt.title('Mean temperature by region')

#cfp.con(f=zon_mean_temps, x=lat, y=alt,ptype=2,lines=False)
```

Separate latitude sections

```python
lat_array = np.array([lat] * 25)

sec_1 = MA.masked_greater(lat_array,-30)
sec_2 = MA.masked_outside(lat_array, -30, 0)
sec_3 = MA.masked_outside(lat_array, 0, 30)
sec_4 = MA.masked_less(lat_array, 30)
'''
cfp.gopen(rows=1, columns=4, bottom=0.2)
cfp.gpos(1)
cfp.con(f=sec_1, x=lat, y=alt[:25], ptype=2,lines=False)
cfp.gpos(2)
cfp.con(f=sec_2, x=lat, y=alt[:25], ptype=2,lines=False)
cfp.gpos(3)
cfp.con(f=sec_3, x=lat, y=alt[:25], ptype=2,lines=False)
cfp.gpos(4)
cfp.con(f=sec_4, x=lat, y=alt[:25], ptype=2,lines=False)
cfp.gclose()'''
```

```python
#make test mask for section and height
h1_s1_temp = MA.masked_where(np.logical_or(height_1.mask,sec_1.mask), zon_temp)
h1_s1_mean = MA.mean(h1_s1_temp)
print(h1_s1_mean)

cfp.con(f=h1_s1_temp, x=lat, y=alt[:25], ptype=2,lines=False)
```

```python
#make array of conditions defining each area
heights = [height_1.mask, height_2.mask, height_3.mask]
sections = [sec_1.mask, sec_2.mask, sec_3.mask, sec_4.mask]
conditions = []
for h in heights:
    for s in sections:
        conditions.append([h,s])
cond = np.asarray(conditions)
#print(cond.shape)
#print(cond[0,:,:,:])
```

```python
#loop over conditions to get mean for each area
law_temps = np.empty((25,73))
for i in range(12):
    temp_values = MA.masked_where(np.logical_or(*cond[i,:,:,:]),zon_temp) #extract temp values in section
    mean = int(np.mean(temp_values)) #get mean over this section
    #print(mean)
    law_temps = np.where(temp_values.mask, law_temps, mean)

law_temps2 = MA.masked_where(zon_press < 25000, law_temps)
```

```python
plt.pcolormesh(lat,alt[:25],law_temps2)
#average over axes 0:time and 3:longitude to get zonal mean
plt.colorbar()

plt.xlabel('Latitude')
plt.ylabel('Model height')
plt.title('Mean temperature by region')
```
