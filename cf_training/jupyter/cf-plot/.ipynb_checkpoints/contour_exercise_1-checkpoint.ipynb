```python
# Import the cf and cf-plot packages
import cf
import cfplot as cfp
```

```python
# Read in the file ncas_data/data1.nc
f = cf.read('~/blueberry_11_19/cf_training/jupyter/ncas_data/data1.nc')
```

```python
# Examine the data to see what fields are in the dataset
print(f)
```

```python
# Select the temperature field and make this selection a single field rather than a list
# Hint use [0] on your selection
temp_data = f[6]
```

```python
# Extract the 1000mb slice of the temperature field
# and make a contour plot of this
temp_1000 = temp_data.subspace(pressure=1000)
cfp.con(temp_1000)
```

```python
# Use cfp.mapset to select plotting of the region 60 to 100 in longitude and 0 to 40 in latitude
# and make a contour plot of the 1000mb temperature
cfp.mapset(60,100,0,40)
cfp.con(temp_1000)
```

```python
# Change the colour scale to 'magma' and remake the contour plot
cfp.cscale('magma')
cfp.con(temp_1000)
```

```python
# Use the cf-python subspace method to make a new variable covering the India region, 
# 60 to 100 in longitude and 0 to 40 in latitude
india = temp_1000.subspace(latitude=cf.wi(0,40),longitude=cf.wi(60,100))
```

```python
# Make a contour plot of this new variable
cfp.con(india)
```
