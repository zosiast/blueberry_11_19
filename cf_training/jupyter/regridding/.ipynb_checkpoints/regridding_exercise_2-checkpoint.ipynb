```python
# Import the cf and cfplot packages
import cf
import cfplot as cfp
```

```python
# Read in ncas_data/model_precip_DJF_means.nc (use the read_field function)
data = cf.read('~/blueberry_11_19/cf_training/jupyter/ncas_data/model_precip_DJF_means.nc')[0]
```

```python
# Inspect the field - it is global in extent
print(data)
```

```python
# Read in ncas_data/alpine_precip_DJF_means.nc (use the read_field function)
data_reg = cf.read('~/blueberry_11_19/cf_training/jupyter/ncas_data/alpine_precip_DJF_means.nc')[0]
```

```python
# Inspect the field - it is a regional field on a projection coordinate system
print(data_reg)
```

```python
# Regrid the regional field to the grid of the global one using the 'bilinear' regridding method
# - unmapped areas will be masked
reg_glob = data_reg.regrids(data, method='bilinear')
```

```python
# Inspect the result
print(reg_glob)
```

```python
# Take the difference between the regridded field and the global data
diff = data - reg_glob
```

```python
# Set the plotting region to approximately over the Alps
cfp.mapset(0,20,30,70)
```

```python
# Create a countour plot of the difference
cfp.con(diff)
```

```python
# Regrid the regional field to the grid of the global one using the 'patch' regridding method
reg_patch = data_reg.regrids(data, method='patch')
```

```python
# Inspect the result
print(reg_patch)
```

```python
# Take the difference between the two  regridded fields
diff_2 = reg_patch - reg_glob
```

```python
# Create a countour plot of the difference
cfp.con(diff_2, lines=False)
```
