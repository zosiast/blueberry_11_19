```python
import netCDF4 as nc
import pylab as plt
import numpy as np
from mpl_toolkits.basemap import Basemap, shiftgrid
import cf
import cfplot as cfp
```

```python
data = nc.Dataset('/shared/netscratch/ptg21/UM_data/xltln/xltln_ch4.nc')
#extract variables to arrays
ch4 = data.variables['ch4']
lat = data.variables['latitude'][:]
lon = data.variables['longitude'][:]

ch4, lon = shiftgrid(180.,ch4,lon,start=False)
```

```python
plt.figure(dpi=150)
map = Basemap(projection='robin', lon_0=0.)
map.drawcoastlines()


lonx,laty = np.meshgrid(lon,lat)
x,y=map(lonx,laty)

cs = map.pcolormesh(x,y,ch4[0,0,:,:]*28.8/16.*1e9)

plt.colorbar()
plt.show()
```
